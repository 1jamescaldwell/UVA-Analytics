A quick script for generating the SAT scores by major for Environmental Science Academic Program Review


```python
import numpy as np
import pandas as pd

graduate_list = pd.read_excel('APR Env Sci.xlsx',sheet_name='Graduates')
SAT_scores = pd.read_excel('APR Env Sci.xlsx',sheet_name='Graduate SAT scores')
```

```python
merged = pd.merge(graduate_list, SAT_scores, on='Student System ID', how='inner')

# Rename VERB to ERWS and MATH to MTSS so consistent names
merged['Test Component'] = merged['Test Component'].replace({
    'VERB': 'ERWS',
    'MATH': 'MTSS'
})
```

```python
summary = merged.pivot_table(
    index="Major (Collapsed exc Interdisciplinary)",
    columns="Test Component",
    values="Score",
    aggfunc="mean"
)

summary.head(10)
```
